```python
import numpy as np
import scipy
from scipy.sparse import csr_matrix

num_A = 10
num_B = 50
graph = np.zeros((num_A + num_B, num_B + num_A), dtype=int)
for a, b in adjacency_list:
    graph[a, num_A + b] = 1
    graph[num_A + b, a] = 1
```

```python
graph
```

```python
num_components, labels = scipy.sparse.csgraph.connected_components(csr_matrix(graph), False)
order = scipy.sparse.csgraph.breadth_first_order(graph, 0, directed=False)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def show_graph_with_labels(adjacency_matrix):
    rows, cols = np.where(adjacency_matrix == 1)
    edges = zip(rows.tolist(), cols.tolist())
    gr = nx.Graph()
    gr.add_edges_from(edges)
    nx.draw(gr, node_size=5, with_labels=True)
    plt.show()
```

```python
show_graph_with_labels(graph)
```
